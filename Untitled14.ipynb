```python
from pyecharts.charts import Pie, Bar, Line,Tab
from pyecharts.globals import ThemeType
from pyecharts import options as opts
from pyecharts.globals import JsCode
pie_style = {
    'normal': {
        'borderRadius': 10,
        'borderWidth': .5,
        'borderColor': 'auto',
    }
}
def draw_pie(data_columns='户型'):
    data_pair=list(data[data_columns].value_counts().items())

    pie = (Pie(init_opts=opts.InitOpts(theme=ThemeType.CHALK, width='1200px', height='600px'))
        .add("户型",
            data_pair=data_pair,
            radius=["25%", "45%"], center=["50%", "50%"],
            itemstyle_opts=pie_style
            )
       
        .set_global_opts(title_opts=[
                dict(text=data_columns, top='50%', left='48%'),
                
            ],
            legend_opts=opts.LegendOpts(is_show=False))
        .set_series_opts(label_opts=opts.LabelOpts(formatter='{b}:{c}\n\n占比:{d}%',
                        font_weight='bold', color='auto'
                        ))
        )
    return pie
draw_pie().render('./render.html')
```

```python

tab = Tab()
for i in data.columns.tolist():
    item = draw_pie(data_columns=i)
    tab.add(item, i)
tab.render('./render.html')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def sns_bar(x, y):
    # 创建一个1行2列的子图布局
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))
    
    # 在第一个子图上绘制条形图
    sns.barplot(x=x, y=y, data=data, ax=axs[0])
    axs[0].set_title(f'{y}与{x}的关系（条形图）')
    
    # 在第二个子图上绘制箱型图
    sns.boxplot(x=x, y=y, data=data, ax=axs[1])
    axs[1].set_title(f'{y}与{x}的关系（箱型图）')
    
    # 调整子图间距
    plt.tight_layout()
    plt.show()

# 调用函数，注意参数名称要与数据框中的列名一致

sns_bar('区', '租价：元/月')
```

```python
from pyecharts import options as opts
from pyecharts.charts import WordCloud, Tab
from pyecharts.globals import ThemeType

def generate_wordcloud_for_each_column(data):
    # 创建一个Tab对象，用于存放多个词云图
    tab = Tab()
    
    # 遍历DataFrame的每一列
    for column in data.columns:
        # 获取当前列的值计数
        dp_cloud = list(data[column].value_counts().items())
        
        # 创建词云图对象，并添加数据
        wordcloud = (
            WordCloud(init_opts=opts.InitOpts(theme=ThemeType.DARK))
            .add("", dp_cloud, word_size_range=[15, 60])
            .set_global_opts(title_opts=opts.TitleOpts(title=f"WordCloud-{column}"))
        )
        
        # 将词云图添加到Tab对象中
        tab.add(wordcloud, column)
    return tab


# 调用函数并渲染到HTML文件
tab = generate_wordcloud_for_each_column(data)
tab.render('./render.html')
```
